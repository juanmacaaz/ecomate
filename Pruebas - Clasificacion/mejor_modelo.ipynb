```python
import os
import numpy as np
import cv2
import time
import json
import matplotlib.pyplot as plt
import albumentations as A

from IPython.display import HTML, display # Para graficar tablas
import tabulate
```

```python
MAIN_PATH = 'Final_dataset_small'

# Calcula el número de imágenes por clase en el conjunto de entrenamiento

train_folders = os.listdir(os.path.join(MAIN_PATH, 'train'))
train_images = []

for folder in train_folders:
    train_images.append(len(os.listdir(os.path.join(MAIN_PATH, 'Train', folder))))

# Calcula el número de imágenes por clase en el conjunto de validación

val_folders = os.listdir(os.path.join(MAIN_PATH, 'Validation'))
val_images = []

for folder in val_folders:
    val_images.append(len(os.listdir(os.path.join(MAIN_PATH, 'Validation', folder))))

# Muestra una tabla greficamente con el número de imágenes por clase en el conjunto de entrenamiento y validación
# el % de imágenes por clase en el conjunto de entrenamiento y validación y el total de imágenes

plt.figure(figsize=(10, 5))
plt.bar(train_folders, train_images, label='Entrenamiento')
plt.bar(val_folders, val_images, label='Validación')
plt.title('Número de imágenes por clase en el conjunto de entrenamiento y validación')
plt.xlabel('Clase')
plt.ylabel('Número de imágenes')
plt.legend()
plt.show()

# Muestra los mismos datos que la gráfica anterior pero en forma de tabla

headers = ['Clase', 'Entrenamiento', '%', 'Validación', '%', 'Total']
table = []

for i in range(len(train_folders)):
    table.append([train_folders[i], train_images[i], round(train_images[i] / sum(train_images) * 100, 2), val_images[i], round(val_images[i] / sum(val_images) * 100, 2), train_images[i] + val_images[i]])

table.append(['Total', sum(train_images), 100, sum(val_images), 100, sum(train_images) + sum(val_images)])

display(HTML(tabulate.tabulate(table, headers, tablefmt='html')))
```

```python
# Muestra 10 imágenes de cada clase del conjunto de entrenamiento
# recorta las imágenes a 224x224 para que se puedan visualizar correctamente

plt.figure(figsize=(5*3, 7))

for i in range(len(train_folders)):
    plt.subplot(2, 5, i + 1)
    plt.title(train_folders[i], fontsize=14)
    plt.axis('off')
    # Quita el espaciado entre las imágenes
    plt.subplots_adjust(wspace=0, hspace=0)
    # Quita los ejes
    plt.gca().axes.get_xaxis().set_visible(False)
    img = cv2.imread(os.path.join(MAIN_PATH, 'Train', train_folders[i], os.listdir(os.path.join(MAIN_PATH, 'Train', train_folders[i]))[5]))
    # Redimensiona la imagen a 500x500 manteniendo la relación de aspecto
    img = cv2.resize(img, (500, 500))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)

plt.show()
```

```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator

# Creación de los generadores de imágenes

def transform(image):
    transform = A.Compose([
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.RandomBrightnessContrast(p=0.5)
    ])
    return transform(image=image)['image']
    

train_datagen = ImageDataGenerator(
    rescale=1./255,
    preprocessing_function=transform
)

train_generator = train_datagen.flow_from_directory(
    os.path.join(MAIN_PATH, 'Train'),
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=True
)

val_datagen = ImageDataGenerator(rescale=1./255)

val_generator = val_datagen.flow_from_directory(
    os.path.join(MAIN_PATH, 'Validation'),
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)
```

```python
def train_model(name, model):

    name = name+'_no_freeze_flatten'

    callbacks = [
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=10,
            verbose=1,
            restore_best_weights=True
        ),
        tf.keras.callbacks.ModelCheckpoint(
            filepath='models/' + name + '.h5',
            monitor='val_loss',
            save_best_only=True,
            verbose=1
        )
    ]

    # AdamW optimizer

    model.compile(optimizer=tf.keras.optimizers.Adam(amsgrad=True),
                    loss='categorical_crossentropy',
                    metrics=['accuracy'])

    history = model.fit(
        train_generator,
        epochs=5,
        validation_data=val_generator,
    )

    

    # Unfreeze all layers

    for layer in model.layers:
        layer.trainable = True

    # Low learning rate

    model.compile(optimizer=tf.keras.optimizers.Adam(amsgrad=True, learning_rate=0.0001),
                    loss='categorical_crossentropy',
                    metrics=['accuracy'])
    
    time_start = time.time()

    history = model.fit(
        train_generator,
        epochs=1000,
        validation_data=val_generator,
        callbacks=callbacks
    )
    
    time_end = time.time()

    data = {}
    data['name'] = name
    data['time'] = time_end - time_start
    data['history'] = history.history
    data['best_val_loss'] = min(history.history['val_loss'])
    data['best_val_acc'] = max(history.history['val_accuracy'])
    data['best_train_loss'] = min(history.history['loss'])
    data['best_train_acc'] = max(history.history['accuracy'])
    data['params'] = model.count_params()
    data['trainable_params'] = len(model.trainable_weights)
    data['non_trainable_params'] = len(model.non_trainable_weights)

    # Elimina el modelo de la memoria de video
    del model
    return data

def save_model_data(data):
    with open('models/' + data['name'] + '.json', 'w') as f:
        json.dump(data, f)

def plot_model_data(data):
    plt.figure(figsize=(10, 5))
    plt.title(data['name'])
    plt.plot(data['history']['accuracy'], label='Entrenamiento')
    plt.plot(data['history']['val_accuracy'], label='Validación')
    plt.xlabel('Época')
    plt.ylabel('Precisión')
    plt.legend()
    plt.show()

    plt.figure(figsize=(10, 5))
    plt.title(data['name'])
    plt.plot(data['history']['loss'], label='Entrenamiento')
    plt.plot(data['history']['val_loss'], label='Validación')
    plt.xlabel('Época')
    plt.ylabel('Pérdida')
    plt.legend()
    plt.show()
```

```python
simple_mlp = tf.keras.models.Sequential([
        tf.keras.layers.Flatten(input_shape=(224, 224, 3)),
        tf.keras.layers.Dense(512, activation='relu'),
        tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
    ])

simple_mlp_data = train_model('simple_mlp', simple_mlp)
print(simple_mlp_data)
save_model_data(simple_mlp_data)
plot_model_data(simple_mlp_data)
```

```python
simple_cnn = tf.keras.models.Sequential([
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
        tf.keras.layers.MaxPooling2D(2, 2),
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
        tf.keras.layers.MaxPooling2D(2, 2),
        tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
        tf.keras.layers.MaxPooling2D(2, 2),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(512, activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
    ])

simple_cnn_data = train_model('simple_cnn', simple_cnn)
save_model_data(simple_cnn_data)
plot_model_data(simple_cnn_data)
```
